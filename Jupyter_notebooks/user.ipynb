# CarCarPro API Schema

                    USER  


1) Create New User

- `/api/v1/user/`

    - `POST` : Register the user


        - **REQUEST BODY**
           
                    ```json    
                    {             
                        
                        "username": "jestin",
                        "password" : "jes@1234",
                        "first_name": "jestin",
                        "last_name": "joy",
                        "email": "tariq.m@dci-student.org",
                        "date_joined": "2024-08-29T21:59:06.414998Z"
                    }
                    ```
               

        -  Response Body

            - Status code: `201 Created`

            ```json
                {
                    
                    "id": 2,
                    "username": "jestin",
                    "first_name": "jestin",
                    "last_name": "joy",
                    "email": "tariq.m@dci-student.org",
                    "date_joined": "2024-09-02T07:54:58.539188Z"

                }
            ```
        -  Response Body

            - Status code: `400 Bad Request`

            ```json
                {
                    "username": [
                        "A user with that username already exists."
                    ]
                }
            ```

2) Login

- `api/v1/auth/login/`
    - `POST` :Login

        - Request Body
            ```json
                {
                "username": "jestin",
                "password" : "jes@1234"
                }
            ```    

        - Response Body
            
            - Status Code : `200 OK`

            ```json
                {
                    "key": "e6a69f8cd1712ff9f7039f0db4ba193f66527fbb"
                }
            ```
            
        - Response Body

            - Status Code : `400 Bad Request`

            ```json
                {
                "non_field_errors": [
                    "Unable to log in with provided credentials."
                ]
                }
            ```


2) List Users

- `api/v1/user/`
    - `GET`: list users

        - ** Headers **
                - Authorization Token


            - Response Body

                - Status code: `200 OK`

                ```json
                    {
                         {
                            "id": 2,
                            "username": "jestin",
                            "first_name": "jestin",
                            "last_name": "joy",
                            "email": "tariq.m@dci-student.org",
                            "date_joined": "2024-09-02T07:54:58.539188Z"
                        },
                        {
                            "id": 1,
                            "username": "admin",
                            "first_name": "",
                            "last_name": "",
                            "email": "",
                            "date_joined": "2024-08-22T17:03:15.197565Z"
                        }
                    }
                ```   

            -  Response Body

                - Status Code: `403 Forbidden`     

                ```json
                    {
                    "detail": "You don't have permission to perform this action."
                    }

            -  Response Body

                - Status Code: `401 Unauthorized`

                ```json
                    {
                    "detail": "Invalid token."
                    }

- `api/v1/auth/logout/`

    - `POST` : LOGOUT

        - Request Body

            ```json
            {
            "username": "admin",
            "password" : "admin"
            }
            ```

        - Response Body

            - Status Code : `200 OK`

            ```json
                {
                "detail": "Successfully logged out."
                }
