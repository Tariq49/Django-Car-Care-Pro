2) Update assigned work status

- `api/v1/mechanic/requests/9/`
    - `PUT`: update work status

        - ** Headers **
            - Authorization Token

        - Request Body

        ```json
          {
            "mechanic_update": "Replaced the oil filter and checked the brakes.",
            "status": "Completed",
            "completed_date": "2024-08-30T12:34:56Z"  
        }
        ```

        - Request Body

        ```json
        {
        "mechanic_update": "Replaced the oil filter and checked the brakes.",
        "status": "In Progress",
        "completed_date": null 
        }
        ```

        - Request Body
        ```json

            {
            "mechanic_update": "Replaced the oil filter and checked the brakes.",
            "status": "In Progress",
            "completed_date": "2024-08-30T12:34:56Z"
            }
        ```

        - Response Body
            - Status Code : `400 Bad Request`

        ```json
            {
            "error": "You can only set completed_date when status is Completed."
            }
        ```
