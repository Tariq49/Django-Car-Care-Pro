             CUSTOMER

1) Add Customer

- `api/v1/customer/`

    - `POST` : Add customer

        - ** Headers **
            - Authorization Token


        - Request Body
            ```json 
                {
                "user": 2,
                "contact_number": "883654321",
                "address": "dertzhenbecker staße 16",
                "gender": "M",
                "profile_image": null
                }
            ```    
        - Response Body
            - Status Code : `201 Created`

            ```json
                {
                    "id": 1,
                    "user": 2,
                    "contact_number": "883654321",
                    "address": "dertzhenbecker staße 16",
                    "gender": "M",
                    "profile_image": null
                }
            ```    

        - Response Body
            - Status Code : `401 Unauthorized`

            ```json
                {
                "detail": "Invalid token."
                }  
            ```    
        - Response Body
            - Status Code : `400 Bad Request`
            
            ```json
                {
                "user": [
                    "Invalid pk \"10\" - object does not exist."
                ]
            }
            ```
            ```json
                {
                "error": "This user already has a customer record."
            }
            ```


2) List Customer

- `api/v1/customer/`
    - `GET`: list customer details

        - ** Headers **
            -  Authorization Bearer Token

        - Response Body
            - Status Code : `200 OK`

            ```json
                {
                "id": 1,
                "user": 3,
                "contact_number": "883654321",
                "address": "dertzhenbecker staße 16",
                "gender": "M",
                "profile_image": null
                }
            ```



        - Response Body
            - Status Code : `200 OK`

            ```json
                {
                "message": "No customers found for this user. Please ensure you are using the correct token or check if the user has any associated customers."
                }  
            ```  

        -  Response Body

            - Status Code: `401 Unauthorized`

            ```json
                {
                "detail": "Invalid token."
                }    
            ```     

        - Response Body

            - Status Code : `403 Forbidden`

        ```json
            {
            "detail": "You do not have permission to perform this action."
            }
        ```

        - Response Body
