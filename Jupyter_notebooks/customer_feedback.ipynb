                                CUSTOMER FEEDBACK


1) Submit feedback for received service

- `api/v1/customer/feedback/`

    - `POST`: submit  feedback

        - ** Headers **
            - Authorization Bearer Token

        - Request Body

            ```json
            {
                "service_id": 8,
                "rating": 5,
                "comments": "The service was excellent!"
            }
            ```

            - Response Body
                - Status Code : `201 Created`

            ```json
            {
            "id": 1,
            "service_id": 8,
            "rating": 5,
            "comments": "The service was excellent!",
            "customer_name": "meenu jtin",
            "service_name": "Oil Change",
            "problem_description": "Oil Change",
            "mechanic_name": "jestin joy"
            }
            ```

2) List Feedback

- `api/v1/customer/feedback/`

    - `GET` : List feedbacks 

        - ** Headers **
            - Authorization Bearer Token

            - Response Body
                - Status Code : `200 OK`

            ```json

            [
                {
                    "id": 1,
                    "service_id": 8,
                    "rating": 5,
                    "comments": "The service was excellent!",
                    "customer_name": "meenu jtin",
                    "service_name": "Oil Change",
                    "problem_description": "Oil Change",
                    "mechanic_name": "jestin joy"
                }
            ]   
            ```

            
            - Response Body
                - Status Code : `200 OK`

            ```json
                {
                "detail": "No feedback found for your services."
                }
            ```    

            - Response Body
                - Status Code : `403 Forbidden`

            ```json
               {
                "detail": "You do not have permission to perform this action."
                } 
            ```

            

            
3) Update feedback

- `api/v1/customer/feedback/detail/`
    - `PUT`: Delete feedback

        - ** Headers **
            - Authentication Bearer Token

            - Request Body

            ```json
                {
                "id": 1, 
                "rating": 2,
                "comments": "poor service"
                } 
            ```

            - Response Body

                - Status Code : `200 OK`

            ```json

                {
                "id": 1,
                "service_id": 8,
                "rating": 2,
                "comments": "poor service",
                "customer_name": "meenu jtin",
                "service_name": "Oil Change",
                "problem_description": "Oil Change",
                "mechanic_name": "jestin joy"
                }
            ``` 

4) Delete feedback

- `api/v1/customer/feedback/detail/`
    - `DELETE`: Delete feedback
        - **  Headers **
            - Authorization Bearer Token

        - Request Body

        ```json
            {
            "id": 1
            }
        ```   

       - Response Body

        - Status Code : `204 No Content`

        ```json
            {
            "detail": "Feedback deleted successfully."
            }
        ```
