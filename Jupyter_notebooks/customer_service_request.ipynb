                    CUSTOMER SERVICE REQUESTS


1) Submit Service Request

- `api/v1/customer/service/request/`
    - `POST` : submit a request

        - ** Headers **
            - Authorization Bearer Token

    - Request Body

    ```json
        {
        "customer": 3,
        "mechanic":4, 
        "vehicles": [
            {
                "vehicle_no": "987",
                "vehicle_name": "Honda",
                "vehicle_model": "2018",
                "vehicle_brand": "City"
            }
        ],
        "category": "Four_wheeler",
        "problem_description": "Oil Change",
        "service_name": "Oil Change",
        "status": "Pending",
        "due_date": null ,
        "mechanic_update": ""
    }
   ``` 

    - Response Body
        - Status Code : `201 Created`

    ```json

            "customer": 1,

            },

                {
    ```

4) Delete Service Request

- `api/v1/customer/service/request/detail/`

    - `DELETE`: delete service request

        - ** Header **
            - Authentication Bearer Token

        - Request Body

        ```json
        {
            "id":6
        }     
        ``` 

        - Response Body
            - Status Code : `204 No Content`
        ```json
        {
        "message": "Service request deleted successfully."
        }            
        ```
        
        - Response Body
            - Status Code : `404 Not Found`
        ```json
            {
            "error": "Service Request not found or not authorized"
            } 
        ```
